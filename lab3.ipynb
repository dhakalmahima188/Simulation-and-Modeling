```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
a = 13
c = 37
m = 256
N = 200
k = 16
seed = 83
alpha = 0.99

# Generate pseudo-random sequence
x = np.zeros(N)
x[0] = seed
for i in range(1, N):
    x[i] = (x[i-1] * a + c) % m

# Plot the histogram
plt.hist(x, bins=np.arange(0, m+1, m/k), facecolor='g')
plt.grid()
plt.axis([0, m, 0, max(np.histogram(x, bins=np.arange(0, m+1, m/k))[0])])
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Histogram of pseudorandom sequence')

# Perform chi-squared test
expected_frequencies = np.ones(k) * (N / k)
observed_frequencies, _ = np.histogram(x, bins=np.arange(0, m+1, m/k))
D = np.sum((observed_frequencies - expected_frequencies)**2 / expected_frequencies)

print(f'Chi-squared statistic (D): {D}')

plt.show()
```
